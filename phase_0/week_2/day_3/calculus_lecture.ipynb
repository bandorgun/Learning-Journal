```python
import sympy as sy
import numpy as np
```

```python
x,y,z = sy.symbols('x y z')
f = 4*x*y + x*sy.sin(z) + x**3 + z**8*y
f
```

```python
x = sy.Symbol('x')
y = sy.Symbol('y')
z = sy.Symbol('z')
```

```python
#Turunan parsial terhadap x
sy.diff(f,x)
```

```python
#Turunan parsial terhadap y
sy.diff(f,y)
```

## Turunan Parsial Numerik

```python
import numpy as np

def f(x,y): 
  return x**2*y+2*x**3*y+y**5

x=np.linspace(1,10)
y=np.linspace(1,10)

#Mengkombinasikan semua kemungkinan dan di pack ke dalam matrix
z=np.array( [[f(i,j) for i in x] for j in y] )

#Ada 2 output do x dan do y
dx,dy=np.gradient(z)
```

```python
z
```

```python
z = []

for i in x:
  tmp = []
  for j in y:
    tmp.append(f(i,j))
  z.append(tmp)
z = np.array(z)
z
```

```python
z.shape
```

```python
dx
```

```python
dy
```

```python
#Plot z,dx,dy yang dianggap sebagai sebuah gambar

fig, ax = plt.subplots(ncols=3,figsize=(17,5))
ax[0].imshow(z)
ax[0].set_title('f(x,y)')
ax[1].imshow(dx)
ax[1].set_title('dx')
ax[2].imshow(dy)
ax[2].set_title('dy')
```

## Luas area lingkaran

```python
from scipy.integrate import trapz
import numpy as np

R=30
x=np.linspace(-R,R,1000)

upper = np.sqrt(R**2 - x**2)
lower = -upper
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(6,6))
ax.plot(x,upper)
ax.plot(x,lower)
```

```python
area = trapz(upper,x) - trapz(lower,x)

print(area)
```

```python
def upper(x):
  return np.sqrt(R**2 - x**2)

def lower(x):
  return -np.sqrt(R**2 - x**2)

area = integrate.quad(upper,-R,R)[0] - integrate.quad(lower,-R,R)[0]
print(area)
```
