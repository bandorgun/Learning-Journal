# DATA CHANGES

```python

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df
```

```python
df = df.drop(columns = ['Total'])
df
```

```python
new_df = df.copy()
new_df
```

```python
new_df['Total'] = new_df.iloc[ : , 4:-2].sum(axis = 1)

new_df
```

```python
# pindahin total ke kiri HP

cols = list(new_df.columns)

new_df = new_df[cols[0:4] + [cols[-1]] + cols[4:-1]]

new_df
```

### Note: kapan axis merepresentasi baris, kapan kolom? (afalin aja dlu)

axes[0] --> indeks untuk baris; axes[1] --> indeks untuk kolom

axis = 1 --> merujuk kepada kolom (pada penjumlahaan menjumlahkan pada baris yang sama); axis = 0 --> (merujuk kepada baris) (pada penjumlahan, menjumlahkan pada kolom yang sama, berlawanan sama intuisi)
